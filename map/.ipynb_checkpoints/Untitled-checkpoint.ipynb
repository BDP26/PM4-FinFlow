```python
import duckdb
import pandas as pd
import numpy as np
import datashader as ds
import datashader.transfer_functions as tf
from datashader.utils import export_image
import PIL.ImageDraw

# ==========================================
# 1. KONFIGURATIONSPARAMETER
# ==========================================
JAHR = 2024
KATEGORIE = "Cetacea"
RASTER_GRAD = 2.0       # Wie groß sind die Quadrate? (Höher = weniger, aber klarere Pfeile)
MIN_PUNKTE = 5          # Mindestanzahl Sichtungen pro Quadrat, um einen Pfeil zu zeichnen
PFEIL_SKALIERUNG = 2.0  # Wie lang sollen die Pfeile dargestellt werden?
LINIEN_DICKE = 2        # Wie dick sind die Pfeile?

OBIS_PATH = "/mnt/shared_data/finflow/obis_raw"

# ==========================================
# 2. DATEN LADEN & VEKTOREN BERECHNEN
# ==========================================
con = duckdb.connect(database=':memory:')
path_obis = f"{OBIS_PATH}/{KATEGORIE}/*/*.parquet"

# Wir berechnen den Schwerpunkt für das 1. Halbjahr (Start) und 2. Halbjahr (Ende)
query = f"""
WITH base AS (
    SELECT 
        decimalLongitude as lon, 
        decimalLatitude as lat,
        (CAST(decimalLongitude / {RASTER_GRAD} AS INT) * {RASTER_GRAD}) as grid_lon,
        (CAST(decimalLatitude / {RASTER_GRAD} AS INT) * {RASTER_GRAD}) as grid_lat,
        CASE WHEN month(eventDate) <= 6 THEN 'START' ELSE 'END' END as phase
    FROM read_parquet('{path_obis}')
    WHERE year(eventDate) = {JAHR}
      AND lon IS NOT NULL AND lat IS NOT NULL
)
SELECT 
    grid_lon, grid_lat,
    AVG(CASE WHEN phase = 'START' THEN lon END) as x1,
    AVG(CASE WHEN phase = 'START' THEN lat END) as y1,
    AVG(CASE WHEN phase = 'END' THEN lon END) as x2,
    AVG(CASE WHEN phase = 'END' THEN lat END) as y2,
    COUNT(*) as anzahl
FROM base
GROUP BY grid_lon, grid_lat
HAVING COUNT(DISTINCT phase) = 2 AND anzahl >= {MIN_PUNKTE}
"""

df_vektoren = con.execute(query).df()

# ==========================================
# 3. VISUALISIERUNG MIT DATASHADER
# ==========================================
# Wir erstellen zuerst eine Heatmap aller Punkte als Hintergrund
query_all = f"SELECT decimalLongitude as lon, decimalLatitude as lat FROM read_parquet('{path_obis}') WHERE year(eventDate) = {JAHR}"
df_all = con.execute(query_all).df()

cvs = ds.Canvas(plot_width=1200, plot_height=700)
agg = cvs.points(df_all, 'lon', 'lat', ds.count())
img = tf.shade(agg, cmap=["#000022", "#000066"], how='log') # Dunkelblauer Hintergrund

# --- PFEILE ZEICHNEN ---
# Da Datashader keine nativen Pfeile hat, nutzen wir PIL auf dem generierten Bild
pil_img = img.to_pil()
draw = PIL.ImageDraw.Draw(pil_img)

# Umrechnung von Koordinaten in Pixel
def get_pixel(lon, lat, w=1200, h=700):
    px = int((lon + 180) * (w / 360))
    py = int((90 - lat) * (h / 180))
    return px, py

for _, row in df_vektoren.iterrows():
    # Startpunkt
    ps = get_pixel(row['x1'], row['y1'])
    
    # Richtung berechnen und skalieren
    dx = (row['x2'] - row['x1']) * PFEIL_SKALIERUNG
    dy = (row['y2'] - row['y1']) * PFEIL_SKALIERUNG
    pe = get_pixel(row['x1'] + dx, row['y1'] + dy)
    
    # Pfeil zeichnen (Linie)
    draw.line([ps, pe], fill="cyan", width=LINIEN_DICKE)
    # Kleiner Punkt am Ende des Pfeils
    draw.ellipse([pe[0]-2, pe[1]-2, pe[0]+2, pe[1]+2], fill="white")

# Bild anzeigen
display(pil_img)
```
